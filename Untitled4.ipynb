```python
#importing packages

import pandas as pd
import networkx as nx
%matplotlib inline
```

```python
#reading the simplified relations file

df = pd.read_csv(r"C:\data\network.csv")
df.head()
```

```python
#Basic drawing

g = nx.from_pandas_dataframe(df, source='PersonOne', target='PersonTwo')
nx.draw(g)
```

```python
#Not very informative, so we need to extract some more information

#Reading a file with all contacts and Id's
df_unique = pd.read_csv(r"C:\data\nodes.csv")

#Creating a list
contacts = list(df_unique.Name.unique())
contacts
```

```python
dict(zip(contacts,contacts))
```

```python
[contact for contact in contacts]
```

```python
#Calculating how many contacts each contact is connected to
[g.degree(contact) for contact in contacts]
```

```python
import matplotlib.pyplot as plt

#Increasing the size of the figure
plt.figure(figsize=(30,30))

#Creating the basic graph
g = nx.from_pandas_dataframe(df, source='PersonOne', target='PersonTwo')

#Changing the layout
layout = nx.spring_layout(g,iterations=50)

contact_size = [g.degree(contact) * 80 for contact in contacts]

#Drawing the nodes
nx.draw_networkx_nodes(g, layout, nodelist=contacts, node_size=contact_size, node_color='lightblue')

#Drawing the "popular" nodes with more than 5 connections
popular_people = [contact for contact in contacts if g.degree(contact) > 5]
nx.draw_networkx_nodes(g, layout, nodelist=popular_people, node_color='orange', node_size=1000)

#Drawing the edges
nx.draw_networkx_edges(g, layout, width=1, edge_color="#cccccc")

#Applying the labels
node_labels = dict(zip(contacts, contacts))
nx.draw_networkx_labels(g, layout, labels=node_labels)
```
